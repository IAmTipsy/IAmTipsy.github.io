```python
import pyglet
import geoplotlib
from geoplotlib.utils import read_csv, BoundingBox, DataAccessObject
import numpy as np
```

```python
#Creating numpy arrays
lat = []
lon = []
lat = np.ndarray(lat)
lon = np.ndarray(lon)
with open("fartkontrol.csv", "rb") as f:
    reader = csv.DictReader(f, delimiter=";")
    for data in reader:
        lat = np.append(lat,np.float64(data["x"]))
        lon = np.append(lon,np.float64(data["y"]))
#Magic          
geo_data_for_plotting = {"lat": lat, "lon": lon}
ndata = geoplotlib.utils.DataAccessObject(geo_data_for_plotting)
geoplotlib.kde(ndata, bw=4, cut_below=1e-4)
geoplotlib.set_bbox(BoundingBox.DK)
#geoplotlib.set_bbox(BoundingBoc.KBH)
geoplotlib.inline()
```

```python
X = []
Y = []
with open("Fartkontroller_sjælland_6mdr.csv", "rb") as f:
    reader = csv.DictReader(f, delimiter=";")
    for data in reader:
        day = int(data['Tilf\xc3\xb8jet'][:2])
        month = int(data['Tilf\xc3\xb8jet'][3:5])
        year = int(data['Tilf\xc3\xb8jet'][6:10])
        date = datetime.date(year,month,day).weekday()
        X.append([int(data["Postnr"]), date])
        if int(data["Bekr\xc3\xa6ftelser"]) < 14:
            Y.append(0)
        else:
            Y.append(1)  
print year, month, day, date
```

```python
lr = LogisticRegression()
lr.fit(X, Y)
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
      intercept_scaling=2, max_iter=50, penalty='12', random_state=None, multi_class="ovr", tol=0.0001)
```
